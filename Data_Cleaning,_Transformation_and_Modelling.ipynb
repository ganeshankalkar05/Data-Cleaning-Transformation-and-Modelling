# Data Cleaning, Transformation and modeling




Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?

A=>Data Loading is the process of connecting to external data sources (like Excel files, SQL databases, web APIs, or cloud services) and importing that data into Power BI's internal storage engine (VertiPaq) or establishing a direct query connection.

It is considered the first step of analysis because:

* Foundation of the Data Model: You cannot visualize, aggregate, or write DAX (Data Analysis Expressions) formulas without first having data present in the environment.

* Performance Dictation: How you load the data (Import mode vs. DirectQuery) dictates the performance and analytical capabilities of your entire dashboard.

* Data Availability: The loading phase determines exactly which tables, columns, and rows will be available for you to analyze. If a critical field isn't loaded, it can't be measured.

Question 2 : Explain the difference between “Load” and “Transform Data” in Power BI.

A=>
* Load: This bypasses any data preparation and pulls the raw data directly into the Power BI data model exactly as it exists in the source. You should only use this option if your data is perfectly clean, correctly formatted, and requires zero modifications (which is rare in real-world scenarios).

* Transform Data: This opens the Power Query Editor, which is Power BI's built-in ETL (Extract, Transform, Load) tool. This allows you to clean and shape your data before it enters the data model. You can use it to remove empty rows, change data types, split columns, merge tables, and filter out irrelevant data

Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset?

A=>
In data modeling, tables are generally divided into these two categories to create a clean, relational structure.

Fact Table:

A fact table contains the quantitative, measurable data (the "facts") of your business, along with foreign keys that link back to dimension tables. Fact tables are typically narrow (few columns) but extremely long (millions of rows) because they record historical events or transactions.

* Example from a Sales Dataset: A Sales table containing Order_ID, Date_Key, Product_Key, Customer_Key, Quantity_Sold, and Total_Revenue.

Dimension Table:
A dimension table contains the descriptive attributes—the "who, what, where, and when"—that give context to your facts. They are used to filter, group, and slice the data in your visuals. Dimension tables are usually wide (many columns) but relatively short (fewer rows) because they contain unique entities.

* Examples from a Sales Dataset: * A Products table containing Product_Key, Product_Name, Category, Brand, and Unit_Cost.

* A Calendar table containing Date_Key, Year, Month, Quarter, and Day_of_Week.

Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?

A=> A Star Schema features a single, central fact table surrounded by directly connected dimension tables (resembling a star). A Snowflake Schema further normalizes those dimension tables into sub-dimensions (e.g., a Products table connected to a separate Categories table)
.Power BI strongly prefers the Star Schema for several key reasons:
* Optimized for the Engine: Power BI's VertiPaq engine is specifically built to compress and query Star Schemas. It handles single-hop relationships (Fact $\leftrightarrow$ Dimension) incredibly fast.
* Faster Query Performance: Snowflake schemas require the engine to traverse multiple relationships (joins) to filter data, which uses more CPU and slows down report rendering. Star schemas minimize these joins.
* Simpler DAX Calculations: Writing DAX measures is much easier and less prone to logic errors in a Star Schema. Complex relationships in a Snowflake schema often require advanced DAX functions (like USERELATIONSHIP or CROSSFILTER) just to get basic totals.
* Intuitive Interface: For the end-user or report builder, a Star Schema provides a much cleaner "Fields" pane. They can easily drag a product category and a sales total onto a canvas without worrying about which sub-table holds the correct filtering context.
